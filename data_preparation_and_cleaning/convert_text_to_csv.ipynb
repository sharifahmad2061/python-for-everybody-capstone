```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('./../data/EXPORTS BY COMMODITIES AND COUNTRIES, (2015-2016).csv')
```

```python
df.head()
```

## Find repeating rows

```python
column_with_values = df['Unnamed: 10']=='VALUE'
```

```python
column_with_values[0] = False
```

```python
column_with_values
```

```python
new_df = df.copy()
```

## Remove 8 repeating rows

```python
for index,value in enumerate(column_with_values):
    if value:
        new_df.drop(list(range(index-8,index+1)),axis=0,inplace=True)
```

```python
new_df[-9:]
```

## Review the data

```python
new_df.sample(10)
```

## find empty rows and delete them

```python
new_df.shape
```

```python
new_df = new_df.dropna(how='all')
```

```python
new_df.shape
```

## replace `--` with NAN

```python
new_df = new_df.replace('--',np.nan)
```

```python
new_df.sample(10)
```

## reset index

```python
new_df = new_df.reset_index().drop('index',axis=1)
```

```python
new_df.head()
```

## Export csv

```python
new_df.to_csv('./../data/2015-2016_cleaned.csv',index=False)
```
