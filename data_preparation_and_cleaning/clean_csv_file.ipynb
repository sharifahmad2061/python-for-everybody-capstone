```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./../data/2018-2019.csv')
```

```python
df.head()
```

## Find repeating rows

```python
column_with_values = df['Unnamed: 10']=='VALUE'
```

```python
column_with_values[0] = False
```

```python
column_with_values
```

```python
column_with_values.value_counts()
```

```python
new_df = df.copy()
```

## Remove 8 repeating rows

```python
for index,value in enumerate(column_with_values):
    if value:
        df.drop(list(range(index-8,index+1)),axis=0,inplace=True)
```

```python
df[-9:]
```

## Review the data

```python
df.sample(15)
```

## find empty rows and delete them

```python
df.shape
```

```python
df = df.dropna(how='all')
```

```python
df.shape
```

## replace `--` with NAN

```python
df = df.replace('--',np.nan)
```

```python
df.sample(10)
```

## reset index

```python
df = df.reset_index().drop('index',axis=1)
```

```python
df.head()
```

## Export csv

```python
df.to_csv('./../data/2018-2019_cleaned.csv',index=False)
```
